# Temperature calibration

Using the peltier device, we calibrate the temperature sensor. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r"../Data/Temperature_calibration.csv")
df.head()
```

```python
plt.scatter(df.Sensor, df.Temperature)
plt.xlabel("Sensor read")
plt.ylabel("Temperature (C)")
```

```python
plt.scatter(df.Peltier, df.Temperature)
plt.xlabel("Peltier voltage (V)")
plt.ylabel("Temperature (C)")
```

These data can then be used to interpret temperature sensor data.

```python
# fit linear sensor data and temperature
p = np.polyfit(df.Sensor, df.Temperature, deg=1)
```

```python
x = np.linspace(-200, 200)
y = p[0] * x + p[1] 
plt.plot(x, y)
```

```python
p
```

We get the following conversion:

$$
T = -0.1 \Delta V_s + 6.8
$$

where $\Delta V_s$ is the voltage difference sensor read. Note that $\Delta V_s$ here is not in physical unit and is dependent on the sensor reader board used.
